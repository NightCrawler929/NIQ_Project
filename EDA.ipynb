```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset = pd.read_csv('en.openfoodfacts.org.products.csv', on_bad_lines='skip', sep='\t')
```

```python
# Set the maximum number of rows displayed
pd.set_option('display.max_rows', 500)

# Set the maximum number of columns displayed
pd.set_option('display.max_columns', None)
```

```python
numerical_columns = ['energy_100g', 'fat_100g', 'saturated-fat_100g']
#Box and whisker plot
for column in numerical_columns:
    plt.figure(figsize=(10, 6))
    plt.boxplot(dataset[column].dropna())
    plt.title(f'Boxplot of {column}')
    plt.xlabel(column)
    plt.show()
```

```python
# Counting the number of products per country
country_counts = dataset['countries'].dropna().value_counts().head(10)

plt.figure(figsize=(10, 6))
country_counts.plot(kind='bar', color='skyblue')
plt.title('Top 10 Countries with Major Products')
plt.xlabel('Country')
plt.ylabel('Number of Products')
plt.xticks(rotation=45)
plt.show()
```

```python
# Counting the number of products per country
country_counts = dataset['countries'].dropna().value_counts().head(10)

plt.figure(figsize=(10, 6))
country_counts.plot(kind='pie', autopct='%1.1f%%')
plt.title('Top 10 Countries with Major Products')
plt.ylabel('')  # Hide the y-label as it's unnecessary in a pie chart
plt.show()
```

```python
brand_counts = dataset['brands'].dropna().value_counts().head(10)

plt.figure(figsize=(10, 6))
brand_counts.plot(kind='bar')
plt.title('Top 10 Brands with Major Products')
plt.xlabel('Brands')
plt.ylabel('Number of Brands')
plt.xticks(rotation=45)
plt.show()
```

```python
# Counting the number of products per country
brand_counts = dataset['brands'].dropna().value_counts().head(10)

plt.figure(figsize=(10, 6))
brand_counts.plot(kind='pie', autopct='%1.1f%%')
plt.title('Top 10 Brands with Major Products')
plt.ylabel('')  # Hide the y-label as it's unnecessary in a pie chart
plt.show()
```

```python
categories_counts = dataset['categories'].dropna().value_counts().head(10)

plt.figure(figsize=(10, 6))
categories_counts.plot(kind='bar', color='green')
plt.title('Top 10 Categories')
plt.xlabel('Categories')
plt.ylabel('Number of Categories')
plt.xticks(rotation=45)
plt.show()
```

```python
# Counting the number of categories
categories_counts = dataset['categories'].dropna().value_counts().head(10)

plt.figure(figsize=(10, 6))
categories_counts.plot(kind='pie', autopct='%1.1f%%')
plt.title('Top 10 Categories')
plt.ylabel('')  # Hide the y-label as it's unnecessary in a pie chart
plt.show()
```

```python
dataset['nutriscore_score'].describe()
```

```python
health_score_columns = ['nutriscore_grade', 'nova_group', 'ecoscore_grade']

colors = ['blue', 'red', 'green']

# Iterate and plot the pie charts
for column, color in zip(health_score_columns, colors):
    counts = dataset[column].dropna().value_counts()
    
    plt.figure(figsize=(8, 8))
    counts.plot(kind='pie', autopct='%1.1f%%', startangle=140)
    plt.title(f'Product Distribution by {column}')
    plt.ylabel('') 
    plt.legend(loc='upper right')
    plt.show()
```

```python
dataset = dataset[dataset['nova_group'] != 375.0].reset_index(drop=True)
```

```python
nova_group_4 = dataset[dataset['nova_group'] == 4]
nova_group_1 = dataset[dataset['nova_group'] == 1]
del dataset

# Top 10 brands & categores contributing to NOVA group 4
top_brands_nova_4 = nova_group_4['brands'].value_counts().head(10)
top_categories_nova_4 = nova_group_4['categories'].value_counts().head(10)

# Top 10 brands & categores contributing to NOVA group 1
top_brands_nova_1 = nova_group_1['brands'].value_counts().head(10)
top_categories_nova_1 = nova_group_1['categories'].value_counts().head(10)

print("Top 10 Brands Contributing to NOVA Group 4 (Highly Processed):")
print(top_brands_nova_4)

print("\nTop 10 Categories Contributing to NOVA Group 4 (Highly Processed):")
print(top_categories_nova_4)

print("\nTop 10 Brands Contributing to NOVA Group 1 (Unprocessed/Minimally Processed):")
print(top_brands_nova_1)

print("\nTop 10 Categories Contributing to NOVA Group 1 (Unprocessed/Minimally Processed):")
print(top_categories_nova_1)
```
